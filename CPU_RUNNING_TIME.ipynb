```python
psnr_scale_2 = [      36.54, 36.49,  36.72,  37.05, 36.58,  37.53, 37.52,  37.62,  37.80]
GPU_Processing_Time =[2,    80,      0.2,   0.009, 4,   0.9,   0.1,    0.9,    0.08]
Name_Of_Models = ["SRCNN","FSRCNN","VDSR","DRCN","LapSRN","DRRN","Ours"]
for k in range(9):
    loc_x = GPU_Processing_Time[k]
    loc_y = psnr_scale_2[k]
    if loc_x == 4 and loc_y == 37.80:
        plt.semilogx(loc_x,loc_y,color='g', marker='s')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=10,color='r')
    else:
        plt.semilogx(loc_x,loc_y,color='b',marker='*')
        plt.text(loc_x+0.01,loc_y,str(Name_Of_Models[k]),fontsize=10)
    
plt.ylim([36,38])
plt.xlim(100,0.0020)
plt.ylabel('PSNR(dB)',fontsize=12)
plt.xlabel('Slow<----------------------Running time(s)---------------------->Fast',fontsize=12)
plt.grid(which="both")
plt.show()
```
