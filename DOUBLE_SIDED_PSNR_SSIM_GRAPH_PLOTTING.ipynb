```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Methods = ['SRCNN',  'VDSR',   'LapSRN',  'IDN',  'CARN',  'IMDN', 'SENext']
PSNR_X2 = [27.40,     28.13,    28.27,    28.46,      28.97,   28.97,  29.22]
SSIM_X2 = [0.7785,    0.8031,   0.8060,   0.8101,    0.8203,  0.8205, 0.8208]
 
fig, ax1 = plt.subplots()
color = 'tab:red'

ax1.set_xlabel('State-Of-the-art-methods')
ax1.set_ylabel('PSNR (dB)', color = color)
ax1.plot(Methods, PSNR_X2, 'o-')
ax1.tick_params(axis ='y', labelcolor = color)
 
# Adding Twin Axes to plot using dataset_2
ax2 = ax1.twinx()
 
color = 'tab:blue'
ax2.set_ylabel('SSIM', color = color)
ax2.plot(Methods, SSIM_X2, 'xr-')
ax2.tick_params(axis ='y', labelcolor = color)
ax1.grid()
ax2.grid()
# Adding title
#plt.title('Ablation study of our SENext Model in Terms of PSNR/SSIM')
# Show plot
plt.show()
```
