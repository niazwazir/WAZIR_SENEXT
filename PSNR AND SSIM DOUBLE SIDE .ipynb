```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Methods = ['SRCNN',  'FSRCNN',  'VDSR',   'DRCN',  'LapSRN',  'DRRN',  'SENext']
PSNR_X2 = [36.66,   37.05,     37.53,    37.63,   37.52,     37.74,   38.04]
SSIM_X2 = [0.9542,     0.9560,     0.9590,    0.9588,   0.9591,     0.9591,   0.9608]
 
fig, ax1 = plt.subplots()
color = 'tab:red'

#ax1.set_xlabel('State Of the Art Methods SET5 SCALE FACTOR 2X')
ax1.set_ylabel('PSNR (dB)', color = color)
ax1.plot(Methods, PSNR_X2,  'o-')
ax1.tick_params(axis ='y', labelcolor = color)
 
# Adding Twin Axes to plot using dataset_2
ax2 = ax1.twinx()
 
color = 'tab:blue'
ax2.set_ylabel('SSIM', color = color)
ax2.plot(Methods, SSIM_X2, 'xr-')
ax2.tick_params(axis ='y', labelcolor = color)
ax1.grid()
ax2.grid()
# Adding title
#plt.title('Ablation study of our SENext Model in Terms of PSNR/SSIM')
# Show plot
plt.show()
```
