```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
Methods =["SRCNN", "FSRCNN", "VDSR", "DRCN", "LapSRN", "DRRN", "MemNet", "ASRN", "IDN", "SRMDNF", "MFEN_S", "CARN", "IMDN","SENext"]
Number_of_Parameters = [57, 32, 620, 1774, 830, 297, 677, 227, 553,1505, 765, 1597, 660, 97]
PSNR = [36.66, 37.05, 37.53, 37.63, 37.52, 37.74, 37.78,37.67, 37.83, 37.79, 38.03, 37.72, 38.00, 38.04]
fig = plt.figure()
ax = fig.add_subplot(111)
# Some random colours: 
colours = range(len(Methods))
ax.scatter(Number_of_Parameters, PSNR, c=colours,s=150)
offset = 50
for x, y, s, country in zip(Number_of_Parameters, PSNR, Number_of_Parameters, Methods):
    ax.text(x+offset, y, country, va='center')
ax.set_xlim(0, 2000)
ax.set_ylim(36.5,38.09)
#ax.set_title('WAZIR PROPOSED METHOD')
ax.set_xlabel('Number of Parameters (K)')
ax.set_ylabel('PSNR (dB) Set5 ×2')
plt.grid(True)
plt.savefig('ABC.pdf')
plt.show()
```
